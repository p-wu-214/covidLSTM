```python
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return asarray(X), asarray(y)
```

```python
n_steps = 7
X, y = split_sequence(china_total['China Total Confirmed Cases'], n_steps)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
X.shape
```

```python
print('Single input');
print(X[0])
print('Single output');
print(y[0])
```

```python
## Size of X
X.shape
```

```python
## Splitting into validation and test sets
number_of_rows = X.shape[0]
n_train = math.ceil(number_of_rows*80/100)
n_test = n_train + math.floor(number_of_rows*10/100)
X_train, y_train, X_val, y_val, X_test, y_test = X[:n_train], y[:n_train], X[n_train:n_test], y[n_train:n_test], X[n_test:], y[n_test:]
```

```python
[X_train.shape, X_val.shape, X_test.shape]
```

```python
X_test.shape
```

```python
X_train.shape[-2:] ## Figure out why it does this
```

```python
tf.keras.backend.clear_session()
simple_lstm_model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(50, input_shape=X_train.shape[-2:]),
    tf.keras.layers.Dense(1)
])
sgd = optimizers.SGD(learning_rate=0.01, momentum=0.98, nesterov=False, name='SGD')
simple_lstm_model.compile(optimizer='sgd', loss='mse', metrics=['mae'])
```

```python
simple_lstm_model.fit(X_train, y_train, epochs=600, batch_size=42, verbose=0,
                      validation_data=(X_val, y_val))
```

```python
simple_lstm_model.evaluate(X_test, y_test, verbose=2)
```
